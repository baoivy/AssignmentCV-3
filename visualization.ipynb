```python
from PIL import Image
from naive_blending import naive_copy
from poissonBlending import poissonBlending
import numpy as np
from IPython.display import display
```

```python
img_src = np.array(Image.open('img/dog.jpg'))
img_target = np.array(Image.open('img/pic1.jpg'))

mask = np.zeros(img_src.shape[:2], dtype=np.uint8)
mask[0:150, 0:150] = 1  

offset = (0, 66)
img_blended = naive_copy(img_src, img_target, mask, offset)

display(Image.fromarray(img_blended))
```

```python
img_src = np.array(Image.open('img/dog.jpg').resize((224, 224)))
img_target = np.array(Image.open('img/pic1.jpg').resize((224, 224)))

mask = np.zeros(img_src.shape[:2], dtype=np.uint8)
mask[0:150, 0:150] = 1  

offset = (0, 66)
img_blended = poissonBlending(img_src, img_target, mask, offset)

display(Image.fromarray(img_blended))
```
